```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load 5 & 30 Year treasury yields
T5 = pd.read_csv("../data/DGS5.csv", header = [0], index_col = [0], parse_dates = [0])
T30 = pd.read_csv("../data/DGS30.csv", header = [0], index_col = [0], parse_dates = [0])
```

### Narrow the data and remove null data rows



```python
T5 = T5.loc['1980-01-01':].dropna()
T30 = T30.loc['1980-01-01':].dropna()
```

```python
T30
```

```python
T5
```

# Merge and Create 50 Day Moving Average   

```python
new = T5.copy()
all = new.merge(T30, left_on='observation_date', right_on='observation_date').copy()
all['5SMA'] = all['DGS5'].rolling(window=50).mean()
all['30SMA'] = all['DGS30'].rolling(window=50).mean()
all.dropna(how='any', axis=0, inplace=True)
```

## Save to file for later use

```python
all.to_csv('data_5_30_SMA.csv')
```
